# Ejercicio1: Crear y usar listas de Python

```python
planets = ['Mercurio', 'Venus', 'Tierra', 'Jupiter', 'Saturno', 'Urano', 'Neptuno']

print( len(planets) )

planets.append('Pluton')

print(f'El numero de planetas actual es {len(planets)} y el ultimo planeta es {planets[-1]}')
```

# Ejercicio 2: Trabajando con datos de una lista

```python
# Lista de planetas
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Neptune']

search = input('Ingresa el nombre de un planeta').capitalize()
planet_index = planets.index(search)

print(f'Planetas m치s cercanos a {search}:')
print(planets[0:planet_index])

print('-'*80)

print(f'Planetas m치s alejados a {search}:')
print(planets[planet_index + 1 :])

```
